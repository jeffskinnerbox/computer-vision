```python
"""
MNIST dataset from "Make Your Own Neural Network" by Tariq Rashid, 2016
https://github.com/makeyourownneuralnetwork/makeyourownneuralnetwork

The handwriten, single digit numbers (zero to nine) are represented in a 28 x 28 pixel gray scale image.

Orginal source is likely http://yann.lecun.com/exdb/mnist/
"""
```

```python
import numpy
import matplotlib.pyplot as plt  # library for plotting arrays

# ensure the plots are inside this notebook, not an external window
%matplotlib inline
```

```python
# load the mnist training data CSV file into a list
test_data_file = open("MNIST Dataset/mnist_test_10.csv", 'r')
test_data_list = test_data_file.readlines()
test_data_file.close()
```

```python
# go through all records in the data set
for record in range(10):
    # split the record by the ',' commas
    all_values = test_data_list[record].split(',')
    
    # first value is the classification, place pixels in array, scale pixels to range 0.01 to 1.00
    classification = all_values[0]
    pixels = ((numpy.asfarray(all_values[1:]) / 255.0 * 0.99) + 0.01).reshape(28,28)
        
    # plot the images
    plt.figure(record)
    plt.title("The Number '" + str(classification) +"'")
    plt.imshow(pixels, cmap='Greys', interpolation='None')
```
