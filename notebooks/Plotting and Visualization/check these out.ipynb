#  xkcd Sketch Drawing Mode
* [Make XKCD-Style Plots From Python](https://hackaday.com/2019/03/07/make-xkcd-style-plots-from-python/)
* [matplotlib.pyplot.xkcd](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.xkcd.html)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 100)
y = (x > 0.5) * (x - 0.5)

plt.xkcd(scale=5, length=400)
plt.xticks([])
plt.yticks([])
plt.ylabel('Downloads of "humor sans" font')
plt.text(0, 0.25, 'Article on xkcd() published')
plt.plot(x, y)
plt.plot([0.3, 0.475], [0.2, 0.025], 'black')
plt.gca().set_aspect(2*9/16)
plt.savefig('xkcd_plot.png', dpi=300)
```
