# Convert RGB Image to Black & White in OpenCV
This Notebook shows you how to convert an RGB image into a black & white (binary) image.

```python
import cv2

picture = "people-walking-above-1"

image = cv2.imread('/home/jeff/Pictures/' + picture + '.png')
cv2.imshow('Color Image',image)

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imwrite('/home/jeff/Pictures/' + picture + '_grayscale.png', gray_image)
cv2.imshow('Gray Image',gray_image)

cv2.waitKey(0)                 # waits forever for user to press any key
cv2.destroyAllWindows()        # closes displayed windows
```

## 1. Read as Grayscale Image
Read the image as a grayscale image If you're reading the RGB image from disk,
then you can directly read it as a grayscale image, like this:

```python
import cv2
import matplotlib.pyplot as plt

im_gray = cv2.imread('/home/jeff/Pictures/' + picture + '.png', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(im_gray)
plt.show()
```

## 2. Convert Image to Binary
You can use [adaptive thresholding](http://opencv.willowgarage.com/documentation/image_processing.html?highlight=adaptivethreshold#cvAdaptiveThreshold)
or [fixed-level thresholding](http://opencv.willowgarage.com/documentation/image_processing.html?highlight=threshold#cvThreshold)
to convert your grayscale image to a binary image.

```python
(thresh, im_bw) = cv2.threshold(im_gray, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
```

## 3. Store to Disk

```python
cv2.imwrite('/home/jeff/Pictures/' + picture + '_BW.png', im_bw)
```
