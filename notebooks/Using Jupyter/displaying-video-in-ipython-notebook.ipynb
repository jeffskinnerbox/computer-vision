# Displaying Video in IPython Notebook

```python
# Import the required modules
%pylab inline 
import cv2
from IPython.display import clear_output
```

```python
# Grab the input device, such as webcam or video file
vid = cv2.VideoCapture("videos/Vincent-Van-Gogh-in-Google-Deep-Dream.mp4")

while(True):
    # Capture frame-by-frame
    ret, frame = vid.read()
    if not ret:
        # Release the Video Device if ret is false
        vid.release()
        # Message to be displayed after releasing the device
        print("Released Video Resource")
        break

    # Convert the image from OpenCV BGR format to
    # matplotlib RGB format to display the image
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # make the out put look like a video display
    axis('off')             # Turn off the axis
    title("Input Stream")   # Title of the window

    # Display the frame within matplotlib
    imshow(frame)
    show()

    # Display the frame until new frame is available
    clear_output(wait=True)

    # keyboard entry of 'q' or esc key to stop
    k = cv2.waitKey(1)
    if chr(k & 255) == 'q' or k == 27:
        vid.destroyAllWindows()
        vid.release()
        break
```
